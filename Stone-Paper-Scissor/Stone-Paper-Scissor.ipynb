```python
import numpy as np
import pandas as pd

choices = np.array(["Stone", "Paper", "Scissors"])

history = []

rounds = int(input("How many rounds do you want to play? "))

for round_no in range(1, rounds + 1):

    print(f"\nRound {round_no}")

    user = input("Choose Stone, Paper, or Scissors: ").capitalize()

    if user not in choices:
        print("Invalid Choice!")
        continue

    computer = np.random.choice(choices)

    if user == computer:
        result = "Draw"

    elif (
        (user == "Stone" and computer == "Scissors") or
        (user == "Paper" and computer == "Stone") or
        (user == "Scissors" and computer == "Paper")
    ):
        result = "Win"

    else:
        result = "Lose"

    history.append([round_no, user, computer, result])

    print("Computer:", computer)
    print("Result:", result)

# Create DataFrame
df = pd.DataFrame(
    history,
    columns=["Round", "Player Choice", "Computer Choice", "Result"]
)

print("\nMatch History")
print(df)

print("\nStatistics")
print(df["Result"].value_counts())

win_rate = (df["Result"] == "Win").mean() * 100
print(f"\nWin Rate: {win_rate:.2f}%")

# Save to CSV
df.to_csv("sps_history.csv", index=False)

print("\nHistory saved to sps_history.csv")
```
